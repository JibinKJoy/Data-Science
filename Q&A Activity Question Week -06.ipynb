### 3. Find out how many countries with respect to the region are available in the dataset

```python
df['Region'].value_counts()
```

### 4.1. On average which region has the highest GDP per capita (current US$)

```python
con1=df.groupby('Region')['GDP per capita (current US$)'].describe()
con1
```

```python
con1.sort_values(by='mean',ascending=False).head()
```

### 4.2. On average which region has the highest International trade: Imports (million US$)

```python
con2=df.groupby('Region')['International trade: Imports (million US$)'].describe()
con2
```

```python
con2.sort_values('mean', ascending=False).head()
```

### 5. Find out which region has more consistent with respect to GDP per capita (current US$)

```python
con1['cv of GDP']=(con1['std']/con1['mean'])*100
con1
```

```python
con1.sort_values(by='cv of GDP')
```

Region with low coeffient of variation (CV) of GDP is more consitent in GDP.
Ocerania is more consistent
