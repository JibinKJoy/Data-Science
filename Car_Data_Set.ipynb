Doing the following operations on ‘cars_data’ dataset:

3. Plot a bar graph of male vs female buyers participated in the sales.
4. Find the top 5 cars based on their sales price.
5. Find the least 5 cars based on their Resell price.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
plt.style.use('seaborn')
```

### 1. Read the dataset to the python environment.

```python
cd=pd.read_csv('cars_data.csv')
cd
```

### 2. Check for the null values present in the dataset.

```python
#Approach 3
cd.isna()
```

### 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
cd['Buyer Gender'].value_counts()
```

```python
#Approach 1
sns.countplot(cd['Buyer Gender'],palette='Accent')
plt.title('Buyer Gender plot', fontsize=15,c='c')
```

```python
#Approach 2
bg_data = {"Buyer Gender": ["Female", "Male"],
        "Count": [5052,4948]}
bg = pd.DataFrame(bg_data, columns=['Buyer Gender', 'Count'])
plt.figure(figsize=(7,7))
plots = sns.barplot(x="Buyer Gender", y="Count", data=bg,alpha=0.50)
# Iterrating over the bars one-by-one
for bar in plots.patches:
    # Using Matplotlib's annotate function and
    # passing the coordinates where the annotation shall be done
    plots.annotate(format(bar.get_height()),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15,xytext=(0,10),
                   textcoords='offset points')
plt.title("No. of Female Buyer vs No. of Male Buyer",c='r')
plt.show()
```

**Insights:**

|No. of female buyers|No. of male buyers|Total no. of buyers|
|-|-|-|
|5052|4948|10,000|
* Number of female buyers > Number of male buyers

### 5. Find the least 5 cars based on their Resell price.

```python
cd_rs=cd.sort_values(by=['Resell Price']).head()
cd_rs
```

```python
cd_rs.loc[: ,['Make','Model','Resell Price']]
```

```python
plt.figure(figsize=(7,10))
plt.subplot(2,1,1)
sns.swarmplot(cd_rs['Resell Price'],cd_rs['Make'],s=15,marker='^')
plt.title('Plot of Resell Price Vs Make',c='b')
plt.grid()
plt.subplot(2,1,2)
sns.swarmplot(cd_rs['Resell Price'],cd_rs['Model'],s=15)
plt.title('Plot of Resell Price Vs Model',c='m')
plt.grid()
```

**Findings**

|**Rank**(from last)|**Make and Model**|
|-|-|
|1|Cheverolet-1500|
|2|GMC Savanna-1500|
|3|Jaguar-XF|
|4|Mitsubushi-3000GT|
|5|Infiniti-G37|
* least 5 cars based on their Resell price can be ranked 
* No car was reselled below 2000
