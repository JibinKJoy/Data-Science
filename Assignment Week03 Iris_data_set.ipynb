# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

## Visualization 1

- Plot used : kde(kernal based density distribution)
- Ploted the four features of each species of iris

```python
df[['SW','SL','PW','PL']].plot.kde()
plt.title('Features distribution across all the Species',fontsize=15)
#plt.grid()
```

### Insights:

1. The width of the sepal(SW) having greater value compared to other and all of having SW between 2 to 4 range.
2. In case of PW and PL, they fall in 2 different ranges respectovely.
3. SL have value ranges from 4 to 8.
4. For all species, PL is as compared to other 3 features.

## Visualization 2
- various features are ploted againist each species using box plot, strip plot, violin plot, swarm plot. 

```python
plt.figure(figsize=(10,20))

plt.subplot(4,1,1)
sns.boxplot(y='SL',x='Classification',data=df)
plt.title('Distribution of SL value for each Species', fontsize=20,c='r')
plt.xlabel('Classification',fontsize=15,c='r')
plt.ylabel('SL',fontsize=15,c='r')

plt.subplot(4,1,2)
sns.stripplot(x='Classification',y='SW',data=df)
plt.title('Distribution of SW value for each Species', fontsize=20,c='b')
plt.xlabel('Classification',fontsize=15,c='b')
plt.ylabel('SW',fontsize=15,c='b')

plt.subplot(4,1,3)
sns.violinplot(x='Classification',y='PL',data=df)
plt.title('Distribution of PL value for each Species', fontsize=20,c='g')
plt.xlabel('Classification',fontsize=15,c='g')
plt.ylabel('PL',fontsize=15,c='g')

plt.subplot(4,1,4)
sns.swarmplot(x='Classification',y='PW',data=df)
plt.title('Distribution of PW value for each Species', fontsize=20,c='c')
plt.xlabel('Classification',fontsize=15,c='c')
plt.ylabel('PW',fontsize=15,c='c')

plt.tight_layout()
```

## Insights:

1. Distribution of **SL value for each Species (Using boxplot)**
    1. The SL of iris virginica is greater compared with other species.
    2. order of SL: iris virginica > iris versicolor > iris setosa (some species of virginica and virsicolor have overlaped SLs.)
    
2. Distribution of **SW value for each Species (Using strip plot)**
    1. Majority of iris setosa have greater SW when combared to other two species.
    2. Except some outliers, most of the iris versicolor and virginica species have almost similar SW.
    3. Some of the Virginica species have more SW than versicolor species.
    
3. Distribution of **PL value for each Species (Using violin plot)**
    1. Iris setosa species have the lowest PL compared to other two.
    2. Order of PL: iris Virginica > iris versicolor > iris setosa
    3. We can observe certain no. of species of virginica and versicolor have similar PL.    
    
4. Distribution of **PW value for each Species (Using Swarm plot)**
    1. order of PW: iris Virginica > iris versicolor > iris setosa
    2. Very few overlap can be evident between virginica and versicolor species, shows some of them have similar PW.

## Visualization 3
- Comparison between Sl & SW, PL & PW

```python
plt.figure(figsize=(7,10))

plt.subplot(2,1,1)
plt.scatter(df['SL'],df['SW'],s=15,c='r')
plt.title('Plot of SL vs SW',fontsize=20,c='m')
plt.xlabel('SL',fontsize=15,c='m')
plt.ylabel('SW',fontsize=15,c='m')

plt.subplot(2,1,2)
plt.scatter(df['PL'],df['PW'],s=15)
plt.title('Plot of PL vs PW',fontsize=20)
plt.xlabel('PL',fontsize=15,c='b')
plt.ylabel('PW',fontsize=15,c='b')

plt.tight_layout()
```

### Insights:

1. Plot of **SL vs SW** (Using Scatter plot)
    1. Even though there is no clear cut idea is evident in this plot, we can observe that species with SW in the range 4 to 4.5 belongs to the species having SL in the range 5 to 6.
    
2. Plot of **PL vs PW** (Using scatter plot)
    1. The species with higher PL have also a higher PW.
